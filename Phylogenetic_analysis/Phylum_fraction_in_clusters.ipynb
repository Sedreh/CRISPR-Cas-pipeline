```python

ct = pd.crosstab(prok.component,prok.group).apply(lambda r: r/r.sum(), axis=1)

ct.plot.bar(stacked=True, cmap=plt.cm.rainbow)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tight_layout()


plt.savefig('/home/is6/work/cas_thesis/16s_extraction/final_round/thesis_results/plot_fraction.png',bbox_inches="tight")
```
